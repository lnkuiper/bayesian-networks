```python
import pandas as pd
```

```python
df = pd.read_csv('bank-additional-full.csv', sep=';')
```

```python
df.head()
```

```python
print(df.keys())
```

```python
df['age'] = df['age'].apply(lambda x: (x >= 18) + (x > 25) + (x > 30) + (x > 45) + (x >= 60))
df['nr.employed'] = df['nr.employed'].apply(lambda x: (x < 5050) + (x < 5150))
df['emp.var.rate'] = df['emp.var.rate'].apply(lambda x: (x < -2.1) + (x < -0.5) + (x < 0.5))
df['cons.price.idx'] = df['cons.price.idx'].apply(lambda x: (x < 92.6) + (x < 93.3) + (x < 93.6) + (x < 94.3))
df['cons.conf.idx'] = df['cons.conf.idx'].apply(lambda x: (x < -45) + (x < -38) + (x < -35))
df['euribor3m'] = df['euribor3m'].apply(lambda x: (x < 3)*1)
```

```python
df.head()
```

```python
df.to_csv('bank-additional-full-binned.csv', index=False)
```
